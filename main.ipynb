```python
def zweischlag(startpunkt, endpunkt, laenge1, laenge2):
    # Berechnen Sie die Gesamtlänge zwischen den beiden Punkten.
    dx = endpunkt[0] - startpunkt[0]
    dy = endpunkt[1] - startpunkt[1]
    gesamtlänge = math.sqrt(dx**2 + dy**2)
    
    # Längen der beiden Linien.
    a = laenge2
    b = laenge1
    
    # Länge der Hypotenuse (Gesamtlänge).
    c = gesamtlänge
    
    # Berechnen Sie den Winkel zwischen den beiden Linien.
    if a + b - 0.1 < c:
        alpha = 0.0
    else:
        alpha = math.acos((b**2 + c**2 - a**2) / (2 * b * c))

    
    # Berechnen Sie die Orientierung der Hypotenuse.
    orientierung = math.atan2(dy, dx)
    
    # Berechnen Sie die Koordinaten des ersten Zwischenpunkts.
    
    zufallszahl = random.random()
    if (zufallszahl > 0):
        zwischenpunkt_x = startpunkt[0] + laenge1 * math.cos(orientierung + alpha) 
        zwischenpunkt_y = startpunkt[1] + laenge1 * math.sin(orientierung + alpha)
    else:
        zwischenpunkt_x = startpunkt[0] + laenge1 * math.cos(orientierung - alpha) 
        zwischenpunkt_y = startpunkt[1] + laenge1 * math.sin(orientierung - alpha)
    
    return (zwischenpunkt_x, zwischenpunkt_y)

def oeffnung(arg1, arg2, arg3, arg4=None):
    if arg4 is None:
        p1 = arg1
        l = arg2
        alpha = arg3
        
        result = p1 + np.array([l * cos(alpha), l * sin(alpha)])
    else:
        p1 = arg1
        p2 = arg2
        l = arg3
        alpha = arg4
        
        orientation = np.arctan2(p2[1] - p1[1], p2[0] - p1[0])
        
        beta = orientation + alpha
        
        result = p2 + np.array([l * cos(beta), l * sin(beta)])
    
    return result

def create_Grid(p1, p2):
    
    vector = np.array([p2[0] - p1[0], p2[1] - p1[1]])
    length = np.linalg.norm(vector)
    
    # length = round(length / 0.01) * 0.01

    # Determine the number of pluralities
    pluralities = int(length)

    # Find the unit vector
    unit_vector = vector / length

    # Create the sequence of points
    points = [p1]
    current_point = p1
    for i in range(1, pluralities + 1):
        current_point = p1 + unit_vector * i
        points.append(current_point)
    return points

def handle_Grid(p1, p2, Thread, Structure):
    new_points = create_Grid(p1, p2)
    for point in new_points:
        existing_warp = None
        for warp in Structure.Warps:
            if all(warp.point == point):
                existing_warp = warp
                break

        if existing_warp:
            if existing_warp.first != Thread.name:
                existing_warp.second = Thread.name
        else:
            new_warp = Warp(point, Thread.name)
            Structure.Warps.append(new_warp)
            
def handle_segments(Structure):

    # Get the points of the last three segments in the last layer
    last_layer_points = Structure.Threads[-1].points[-3:-1]

    # print(last_layer_points)

    # Check if there are points above the lines formed by the last three segments
    tol = 0.1
    for point in last_layer_points:
        x, y = point
        for i in range(len(Structure.Threads[-2].points) - 1):
            x1, y1 = Structure.Threads[-2].points[i]
            x2, y2 = Structure.Threads[-2].points[i + 1]

            

            if x1 <= x <= x2:
                expected_y = y1 + (y2 - y1) * (x - x1) / (x2 - x1)
                if y + tol  < expected_y:
                    # Point is below the line, delete last_layer_points and return False
                    Structure.Threads[-1].points = Structure.Threads[-1].points[:-4]
                    return False
                
    # After successful plausibility check
    handle_Grid(Structure.Threads[-1].points[-4], Structure.Threads[-1].points[-3], Structure.Threads[-1], Structure)
    handle_Grid(Structure.Threads[-1].points[-3], Structure.Threads[-1].points[-2], Structure.Threads[-1], Structure)
    handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)
    return True  # Return True if the segments were successfully handled

def orientation(p1, p2):
    return math.atan2(p1[1] - p2[1], p1[0] - p2[0])

def Weave(Structure, Setup):

    height = Setup.height
    sigma = Setup.sigma
    threadname = 1
    
    if (Setup.noml)%4 != 0:
        raise ValueError("length of weave is not devisible by 4")
        
    # ERSTER FADEN
    Structure.Threads.append(Thread(np.array([0, 0]), "B", threadname))
    threadname += 1
    Structure.Threads[0].points.append(oeffnung(Structure.Threads[0].points[-1], Setup.leng, 0))
    handle_Grid(Structure.Threads[0].points[-2], Structure.Threads[0].points[-1], Structure.Threads[0], Structure)
    for _ in range(int(Setup.noml / 4) +1):
        
        neigung = orientation(Structure.Threads[0].points[-2], Structure.Threads[0].points[-1]) - pi
        
        lengths, angles = [],[]

        for i in range(4):
            lengths.append(Setup.leng * random.gauss(1, sigma))
            angles.append(60 / 360 * 2 * pi * random.gauss(1, sigma / 2))
        angles[0] = -angles[0] - neigung
        angles[3] = -angles[3]

        for i in range(4):
            point = oeffnung(Structure.Threads[0].points[-2], Structure.Threads[0].points[-1], lengths[i], angles[i])
            Structure.Threads[0].points.append(point)
            handle_Grid(Structure.Threads[0].points[-2], Structure.Threads[0].points[-1], Structure.Threads[0], Structure)

    # FOLGENDE FÄDEN
    Threads = Structure.Threads
    for _ in range(height - 1):
        if Threads[-1].kind == 'A':
            Threads.append(Thread(np.array([0, Threads[-1].points[0][1] + Setup.leng * sin(60 * 2 * pi / 360) * 2 * random.gauss(1, sigma)]), 'B', threadname))
            threadname += 1

            lengths = math.sqrt((Threads[-1].points[0][0] - Threads[-2].points[2][0])**2 + (Threads[-1].points[0][1] - Threads[-2].points[2][1])**2)
            while True:
                leng1 = lengths/1.73 * random.gauss(1, sigma)
                leng2 = lengths/1.73 * random.gauss(1, sigma)
                if leng1 + leng2 > lengths:
                    break

            midpoint = np.array(zweischlag(Threads[-1].points[-1], Threads[-2].points[len(Threads[-1].points) + 1], leng1, leng2))
            Threads[-1].points.append(midpoint)
            handle_Grid(Threads[-1].points[-2], Threads[-1].points[-1], Threads[-1], Structure)

            Structure.Threads[-1].points.append(Threads[-2].points[len(Threads[-1].points)])
            handle_Grid(Threads[-1].points[-2], Threads[-1].points[-1], Threads[-1], Structure)

        elif Threads[-1].kind == 'B':
            Structure.Threads.append(Thread(np.array([0, Structure.Threads[-1].points[0][1]]), 'A', threadname))
            threadname += 1

        repetitions = int(Setup.noml / 4) if Threads[-1].kind == 'B' else int(Setup.noml / 4) + 1

        for _ in range(repetitions):
                    while True:
                        leng1, leng2, leng3, ang = simple_rng(Structure, Setup)

                        Structure.Threads[-1].points.append(Structure.Threads[-2].points[len(Structure.Threads[-1].points)])
                        # handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)

                        point = oeffnung(Structure.Threads[-1].points[-1], leng1, ang)
                        Structure.Threads[-1].points.append(point)
                        # handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)

                        midpoint = np.array(zweischlag(Structure.Threads[-1].points[-1], Structure.Threads[-2].points[len(Structure.Threads[-1].points) + 1], leng2, leng3))

                        Structure.Threads[-1].points.append(midpoint)
                        # handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)

                        Structure.Threads[-1].points.append(Structure.Threads[-2].points[len(Structure.Threads[-1].points)])
                        # handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)
                        if handle_segments(Structure):
                            break
        if Structure.Threads[-1].kind == 'B':
            Structure.Threads[-1].points.append(Structure.Threads[-2].points[len(Structure.Threads[-1].points)])
            handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)

            leng3, ang = Setup.leng, 60 / 360 * pi * 2

            point = oeffnung(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], leng3, ang)
            Structure.Threads[-1].points.append(point)
            handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)

                
            leng3 = Setup.leng * random.uniform(0.8, 1.2)
            leng3 = round(leng3)
            ang = random.uniform(0.9 * 60 / 360 * 2 * pi, 1.1 * 60 / 360 * 2 * pi)    
                
            Structure.Threads[-1].points.append(oeffnung(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], leng3, - ang))
            handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)
        elif Structure.Threads[-1].kind == 'A':
            Structure.Threads[-1].points.append(Structure.Threads[-2].points[len(Structure.Threads[-1].points)])
            handle_Grid(Structure.Threads[-1].points[-2], Structure.Threads[-1].points[-1], Structure.Threads[-1], Structure)
        
 
def simple_rng(Structure, Setup):

    print('call rng')
    
    sigma = Setup.sigma
    
    forThread = Structure.Threads[-1]
    forindex = len(forThread.points)
    beforeThread = Structure.Threads[-2]
        
    while True:
        try:
            neigung = orientation(beforeThread.points[forindex-2], beforeThread.points[forindex -1]) #- pi
        except:
            neigung = 0
            
        # omega = math.atan2(beforeThread.points[forindex + 2][1]-beforeThread.points[forindex + 1][1], beforeThread.points[forindex + 2][0] - beforeThread.points[forindex + 1][0])
        neigung = neigung * (1/math.exp(neigung/2.718))

        ang = 60 / 360 * 2 * pi * random.gauss(1, sigma/2) #- neigung

        leng1 = Setup.leng * random.gauss(1, sigma)
        leng1 = round(leng1)

        p1 = oeffnung(Structure.Threads[-2].points[len(Structure.Threads[-1].points)], leng1, ang)
        print(Structure.Threads[-2].points[len(Structure.Threads[-1].points)], leng1, ang)

        length = math.sqrt((p1[0] - beforeThread.points[forindex + 3][0])**2 + (p1[1] - beforeThread.points[forindex + 3][1])**2) / 1.73
        print(p1, beforeThread.points[forindex + 3])
        min_dist = math.sqrt((p1[0] - beforeThread.points[forindex + 2][0])**2 + (p1[1] - beforeThread.points[forindex + 2][1])**2)
        print(min_dist)
        print('')

        # Generiere Längen für leng2 und leng3
        while True:
            leng2 = length * random.gauss(1, sigma)
            leng2 = round(leng2)
            leng3 = length * random.gauss(1, sigma)
            leng3 = round(leng3)

            if leng2 + leng3 > min_dist:
                break
            elif min_dist%1 == 0:
                leng2 = round(random.random() * min_dist / 2)
                leng3 = min_dist - leng2
                break

            if leng2 + leng3 > 1.05 * length *  1.73:
                print(leng1, leng2, leng3, ang, length, beforeThread.points[forindex + 3], p1)
                break
        return leng1, leng2, leng3, ang  

# def simple_rng(Structure, Setup):

#     print('call rng')
    
#     sigma = Setup.sigma
    
#     forThread = Structure.Threads[-1]
#     forindex = len(forThread.points)
#     beforeThread = Structure.Threads[-2]
        
#     while True:
#         try:
#             neigung = orientation(beforeThread.points[forindex-2], beforeThread.points[forindex -1]) #- pi
#         except:
#             neigung = 0
            
#         # omega = math.atan2(beforeThread.points[forindex + 2][1]-beforeThread.points[forindex + 1][1], beforeThread.points[forindex + 2][0] - beforeThread.points[forindex + 1][0])
#         neigung = neigung * (1/math.exp(neigung/2.718))

#         ang = 60 / 360 * 2 * pi * random.gauss(1, sigma/2) #- neigung

#         supposed_length = Setup.leng
#         supposed_length = math.sqrt((beforeThread.points[forindex -1][0] - beforeThread.points[forindex + 2][0])**2 + (beforeThread.points[forindex -1][1] - beforeThread.points[forindex + 2][1])**2) / 4 + math.sqrt((beforeThread.points[forindex -1][0] - beforeThread.points[forindex + 3][0])**2 + (beforeThread.points[forindex -1][1] - beforeThread.points[forindex + 3][1])**2) / 4
#         supposed_length = math.sqrt(1/2 * Polygon(zip([beforeThread.points[forindex + 3][0], beforeThread.points[forindex + 2][0], beforeThread.points[forindex + 1][0], beforeThread.points[forindex ][0] ], [beforeThread.points[forindex + 3][1], beforeThread.points[forindex + 2][1], beforeThread.points[forindex + 1][1], beforeThread.points[forindex ][1] ])).area)

#         leng1 = supposed_length * random.gauss(1, sigma)
#         leng1 = round(leng1)

#         p1 = oeffnung(Structure.Threads[-2].points[len(Structure.Threads[-1].points)], leng1, ang)
#         print(Structure.Threads[-2].points[len(Structure.Threads[-1].points)], leng1, ang)

#         length = math.sqrt((p1[0] - beforeThread.points[forindex + 3][0])**2 + (p1[1] - beforeThread.points[forindex + 3][1])**2) / 1.73
#         print(p1, beforeThread.points[forindex + 3])
#         min_dist = math.sqrt((p1[0] - beforeThread.points[forindex + 2][0])**2 + (p1[1] - beforeThread.points[forindex + 2][1])**2)
#         print(min_dist)
#         print('')

#         # Generiere Längen für leng2 und leng3
#         while True:
#             leng2 = length * random.gauss(1, sigma)
#             leng2 = round(leng2)
#             leng3 = length * random.gauss(1, sigma)
#             leng3 = round(leng3)

#             if leng2 + leng3 > min_dist:
#                 break
#             elif min_dist%1 == 0:
#                 leng2 = round(random.random() * min_dist / 2)
#                 leng3 = min_dist - leng2
#                 break

#             if leng2 + leng3 > 1.05 * length *  1.73:
#                 print(leng1, leng2, leng3, ang, length, beforeThread.points[forindex + 3], p1)
#                 break
#         return leng1, leng2, leng3, ang
    
```

```python
def weft_warp_intersection(point, thread):
    x, y = point
    closest_distance = float('inf')
    closest_changepoints = None

    for i in range(1, len(thread.points)):
        changepoints = thread.points
        x1, y1 = changepoints[i - 1]
        x2, y2 = changepoints[i]

        if (x1 <= x <= x2) or (x2 <= x <= x1):
            distance = abs(x - x1)

            if distance < closest_distance:
                closest_distance = distance
                closest_changepoints = (x1, y1, x2, y2)

    if closest_changepoints is not None:
        x1, y1, x2, y2 = closest_changepoints
        expected_y = y1 + (x - x1) * (y2 - y1) / (x2 - x1)

        if np.isclose(y, expected_y, atol=0.01):
            return 2
        elif y > expected_y:
            return 1
        else:
            return 0

    # raise ValueError("weft warp logic")

# Sort the Warps by the x-component of the point attribute
sorted_warps = sorted(myStructure.Warps, key=custom_sort_key)

# # Iterate through the sorted Warps
# for warp in sorted_warps:
    
#     print(f"{str(warp.point):<40}", end=' ')
    
#     for Thread in myStructure.Threads:
#         print(weft_warp_intersection(warp.point, Thread), end='  ')
#     print(' ')

liste = []

# Iterate through the sorted Warps
for warp in sorted_warps:
    # Extract the x and y coordinates as floats with two decimal places
    x_coordinate = warp.point[0]
    y_coordinate = warp.point[1]
    
    # Create a new list to store information for this Warp
    warp_info = [f"{x_coordinate:.2f}", f"{y_coordinate:.2f}"]

    # Iterate through the Threads and calculate intersections
    for Thread in myStructure.Threads:
        intersection_result = weft_warp_intersection(warp.point, Thread)
        warp_info.append(intersection_result)

    # Append the warp_info list to liste
    liste.append(warp_info)
    
liste
```
